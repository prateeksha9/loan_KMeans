```python
#import required libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
```

```python
#import database
df = pd.read_csv('C:\\Users\\Dell\\Desktop\\test\\loan_data_set.csv')
```

```python
#get some top rows of the database
df.head()
```

```python
#drop Loan_ID and Loan_Status
df = df.drop('Loan_ID', axis=1)
df = df.drop('Loan_Status', axis=1)
```

```python
df
```

```python
#remove Nan Values
nan_value = float("NaN")
df.replace("", nan_value, inplace=True)
df.dropna(subset = ["LoanAmount"],inplace=True)
df.dropna(subset = ["CoapplicantIncome"],inplace=True)
```

```python
#get the datatypes of all the coloumns
df.dtypes
```

```python
#plot Scatter plot
plt.scatter(df.CoapplicantIncome,df.LoanAmount)
```

```python
#From scatter plot we can see that there are 2 cluster.
km = KMeans(n_clusters=3)
km
```

```python
#Here, we are basically creating an array which will predict and fit the two columns in an array 
x_predicted = km.fit_predict(df[['CoapplicantIncome','LoanAmount']])
x_predicted
```

```python
#x_predicted added to the database with a coloumn name "cluster" 
df["cluster"] = x_predicted
df.head()
```

```python
#Plot KMeans Plot
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster ==2]


plt.scatter(df1.CoapplicantIncome,df1.LoanAmount,color="green")
plt.scatter(df2.CoapplicantIncome,df2.LoanAmount,color="red")
plt.scatter(df3.CoapplicantIncome,df3.LoanAmount,color="black")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='+',label='centeroid')
plt.xlabel('LoanAmount')
plt.ylabel('CoapplicantIncome')
plt.legend()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('C:\\Users\\Dell\\Desktop\\test\\loan_data_set.csv')
```

```python
df.head()
```

```python
df = df.drop('Loan_ID', axis=1)
```

```python
df = df.drop('Loan_Status', axis=1)
```

```python
df
```

```python
x=scale(df.ApplicantIncome)
```

```python
plt.scatter(df.ApplicantIncome,df.LoanAmount)
```

```python
nan_value = float("NaN")
df.replace("", nan_value, inplace=True)
df.dropna(subset = ["LoanAmount"],inplace=True)
```

```python
df
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['ApplicantIncome','LoanAmount']])
y_predicted
```

```python
df["cluster"] = y_predicted
df.head()
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1.ApplicantIncome,df1.LoanAmount,color="green")
plt.scatter(df2.ApplicantIncome,df2.LoanAmount,color="red")
plt.scatter(df3.ApplicantIncome,df3.LoanAmount,color="blue")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='+',label='centeroid')
plt.xlabel('LoanAmount')
plt.ylabel('ApplicantIncome')
plt.legend()

```
